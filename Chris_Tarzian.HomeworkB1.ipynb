```python
import time
import pandas as pd
from bs4 import BeautifulSoup

# ! pip install selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
# visit a website
path    = 'chromedriver'
browser = webdriver.Chrome(path)

#choose twitter's explore page URL
link = "https://twitter.com/explore"
browser.get(link)
```

```python
#find element of the search bar
element = '//input[@class="r-30o5oe r-1niwhzg r-17gur6a r-1yadl64 r-deolkf r-homxoj r-poiln3 r-7cikom r-1ny4l3l r-xyw6el r-y0fyvk r-1dz5y72 r-fdjqy7 r-13qz1uu"]'

#click on element
browser.find_element_by_xpath(element).click()
time.sleep(2)

#write on element
browser.find_element_by_xpath(element).send_keys("#MacMiller")
time.sleep(2)

#executing search
browser.find_element_by_xpath(element).send_keys(Keys.RETURN)
time.sleep(2)
```

```python
### use while loop

#creating lists for data frame
tweets = []
authors = []

#creating an initial value for the scroll down function
x = 1000

tweet_count = 0

while tweet_count <10:
        page_source = browser.page_source
        soup = BeautifulSoup(page_source, 'lxml')
        reviews_content = soup.find_all('div', class_='css-1dbjc4n r-eqz5dr r-16y2uox r-1wbh5a2')

        # extract the author and tweet
        for review in reviews_content:

            author = review.find('span', class_='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0').text
            #band-aid solution for unreadable tweets
        
            if tweet_count < 10:
                try:
                    tweet = review.find('div', class_='css-901oao r-18jsvk2 r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0').text

                except:
                    print('unreadable tweet')
                # append to our accumulative lists
                tweets.append(tweet)
                authors.append(author)

                tweet_count += 1

            #scroll down
            browser.execute_script("window.scrollTo(0, "+str(x)+")")

            #increment value to our x variable to continue scrolling down and not stay stationary
            x += 1000

            time.sleep(1)

#gathering data into a df
data = {
    'Tweets':tweets,
    'Authors':authors
}

df = pd.DataFrame(data)

# checking our dataframe
df
```
